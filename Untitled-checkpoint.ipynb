```python
# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)

# To plot pretty figures
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

The base object used in the EHR is a dictionary of patients.  We will mimic that here with 10000 patients.

Two important caveats for data structure. The first entry is the current A1c level. The second entry is whether that value was imputed or is real. That is a binary value, and we only evaluate on points that were observed.
The rest of the covariates are lab values, whether they were observed, and time to last observation. For this task we will only mimic the structure and not perform it exactly.

The number of time points is variable per individual.

The generation model will be an AR(1) will external inputs, guess more like a sarimax if you want to be technical.

```python
# create id numbers
random.seed(42) # for reproducibility
id_numbers=random.choice(np.arange(1,1e5,dtype=np.int32),replace=False,size=[5000])
p=39
observed_amount=.25
```

```python
# min/max length
combined_data=dict()
random.seed(42) # for reproducibility
min_length=12
max_length=100
W=.02*random.randn(p,1)
W[0,0]=.95 # self-decay parameter
innovation_weight=.03
for pat in id_numbers:
    ## number of observations
    num_samples=random.choice(max_length-min_length,size=[1])+min_length
    num_samples=num_samples[0]
    ## random covariates
    X=np.zeros(shape=[num_samples,p])
    X[:,2:]=random.randn(num_samples,p-2)
    ## what is actually observed
    obs=random.rand(num_samples)<observed_amount
    obs[0]=True
    obs[-1]=True
    X[:,1]=obs
    ##
    X[0,0]=random.randn(1)
    for t in range(1,num_samples):
        X[t,0]=np.dot(X[t-1,:],W)+innovation_weight*random.randn(1)
    combined_data[pat]=X
```

Simple processing and prediciton timeline

```python
from sklearn.model_selection import train_test_split
pat_nums, pat_nums_valid=train_test_split(id_numbers,random_state=42)
```

```python
num_features=p
```

```python
def draw_data(batch_size=1,valid=False):
    if valid:
        batch_pat_nums=np.random.choice(pat_nums_valid,size=batch_size,replace=False)
    else:
        batch_pat_nums=np.random.choice(pat_nums,size=batch_size,replace=False)
    batch_data=np.zeros(np.int32([batch_size,max_length,num_features]))
    for i in range(0,batch_size):
        tmp=np.nan_to_num(combined_data[batch_pat_nums[i]])
        batch_data[i,0:tmp.shape[0],:]=tmp.reshape(1,-1,num_features)
        batch_data[i,tmp.shape[0]:,:]=np.repeat(tmp[-1,:].reshape(1,1,num_features),repeats=max_length-tmp.shape[0],axis=1)
        batch_data[i,tmp.shape[0]:,1]=0
    X_batch=batch_data[:,:-1,:]
    y_batch=batch_data[:,1:,0].reshape(batch_size,-1,1)
    o_batch=batch_data[:,1:,1].reshape(batch_size,-1,1)
    return X_batch, y_batch, o_batch
```

Model definition

```python
reset_graph()

n_steps = max_length-1
n_inputs = num_features
n_neurons = 40
n_outputs = 1

data_in = tf.placeholder(tf.float32, [None, n_steps, n_inputs])
y = tf.placeholder(tf.float32, [None, n_steps, n_outputs])
o = tf.placeholder(tf.float32, [None, n_steps, 1])
```

```python
cell = tf.contrib.rnn.BasicRNNCell(num_units=n_neurons, activation=tf.nn.relu)
#cell = tf.contrib.rnn.BasicLSTMCell(num_units=n_neurons, activation=tf.nn.relu)
rnn_outputs, states = tf.nn.dynamic_rnn(cell, data_in, dtype=tf.float32)
```

```python
learning_rate = 1e-4
```

```python
stacked_rnn_outputs = tf.reshape(rnn_outputs, [-1, n_neurons])
stacked_outputs = tf.layers.dense(stacked_rnn_outputs, n_outputs)
output_intermediate = tf.reshape(stacked_outputs, [-1, n_steps, n_outputs])
bias=tf.get_variable('bias',[1],tf.float32)
outputs=1.*output_intermediate+0.*tf.reshape(data_in[:,:,0],[-1,n_steps,1])+bias
```

```python
loss = tf.reduce_sum((.1+o)*tf.square(outputs - y))/tf.reduce_sum(o)
realmse= tf.reduce_sum((.0+o)*tf.square(outputs - y))/tf.reduce_sum(o)
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

Training

```python
sess= tf.InteractiveSession()
```

```python
n_iterations = 1500
batch_size = 10


with sess.as_default():
    #with tf.Session() as sess:
    init.run()
    for iteration in range(n_iterations):
        X_batch, y_batch, o_batch = draw_data(batch_size)
        sess.run(training_op, feed_dict={data_in: X_batch, y: y_batch, o: o_batch})
        if iteration % 50 == 0:
            out = outputs.eval(feed_dict={data_in: X_batch, y: y_batch, o: o_batch})
            mse = realmse.eval(feed_dict={data_in: X_batch, y: y_batch, o: o_batch})
            y_batch = sess.run(outputs, feed_dict={data_in: X_batch})
            if np.isnan(mse):
                break
            if np.isnan(mse):
                break
            print(iteration, "\tloss:", mse)
    
    
    #X_new = time_series(np.array(X_new.reshape(-1, n_steps, n_inputs)))
    #y_pred = sess.run(outputs, feed_dict={X: X_new})

saver.save(sess, "./my_time_series_model")
```
